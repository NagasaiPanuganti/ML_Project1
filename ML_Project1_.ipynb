```
#Visualization of Country-wise COVID-19 confirmed cases (on 22nd March,2020)
```

```
#countries having less than 1500 cases as of 23th March,2020 are considered as others
country=dict()
for i in range(487):
  if(df.iloc[i][-1]<1500):
    country["others"]=df.iloc[i][-1]+country.get("others",0)
  else:
    country[df.iloc[i][1]]=df.iloc[i][-1]+country.get(df.iloc[i][1],0)

#view country dict 
for cnt in country:
  print("Cases:{} in Country:{} ".format(country[cnt],cnt))
```

```
plt.figure(figsize=(15,15))
plt.pie(country.values(),labels=country.keys(),autopct='%1f%%')
plt.title('Confirmed Cases per Country as of 22nd March,2020',size=30)
plt.axis('equal')
plt.legend(country.keys(), loc="lower right",prop={'size' :15} )
```

```
df1=df.drop(["Province/State","Lat","Long"],axis=1)
df1
```

```
df2=df1.isin([0]).sum()
#df2 is a series(1-D)
df2
```

```
column_list=[]
#341 is the 70% of 487, we are checking: if number of zero cases on a date is more than 70%, we are appending that index to the column_list, so that we can drop that particular column.
for i in range(1,62):
  if(df2[i]>341):
    column_list.append(i)

df1.drop(df1.columns[column_list],axis=1,inplace=True)
```

```
df1
#compressed DataFrame, as we dropped some columns
```

```
df1.drop(["Country/Region"],axis=1, inplace=True)
```
